```python
g.resync_bag()
words = set()
hand = tuple(g.make_hand())
print(hand)

for i, direction, line in g.playable_lines():

    for j, l in line.items():
        if j > 0 and l and line[j - 1] or j > 0 and not l and line[j - 1]:
            continue
        pattern = tuple(line[j:])
        coord = (i, j) if direction is ACROSS else (j, i)

        for word in g.word_tree.find_playable(pattern, hand):
            secondaries = list(g.find_secondary_words(word, coord, direction))
            
            if secondaries and all(word in g.all_words for *_, word in secondaries):
                score = g.calc_play_score(word, coord, direction, secondaries)
                if not score:
                    continue
                words.add((word, coord, direction, pattern, tuple(secondaries), score, g.calc_turn_val(word, score, hand)))

words = list(words)
```
